```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re,os,glob,pickle
import datetime
import matplotlib
import warnings
warnings.filterwarnings('ignore')
import yaml,logging
from datetime import timedelta
# importing "math" for mathematical operations 
import math 

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
exited_emp_list = pd.read_csv('Exited_Employee_Data.csv')
## Calculating Years at Company
exited_emp_list['yearsAtCompany'] = pd.to_datetime(exited_emp_list['Exit Date']) - pd.to_datetime(exited_emp_list['Most Recent Hire Date'])
exited_emp_list['yearsAtCompany'] = (exited_emp_list['yearsAtCompany'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)


exited_emp_list['Last Promotion Date'] = np.where(exited_emp_list['Last Promotion Date'].notnull(), exited_emp_list['Last Promotion Date'], datetime.datetime.now())
exited_emp_list['YearsSinceLastPromotion'] = pd.to_datetime(exited_emp_list['Last Promotion Date']) - pd.to_datetime(exited_emp_list['Most Recent Hire Date'])
exited_emp_list['YearsSinceLastPromotion'] = (exited_emp_list['YearsSinceLastPromotion'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)

## Calculating Age from Date of Birth
exited_emp_list['Date of Birth'] = np.where(exited_emp_list['Date of Birth'].notnull(), exited_emp_list['Date of Birth'], datetime.datetime.now())
exited_emp_list['Age'] = pd.to_datetime('now') - pd.to_datetime(exited_emp_list['Date of Birth'])
exited_emp_list['Age'] = (exited_emp_list['Age'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)


active_emp_list = pd.read_csv('Active_Employee_Data.csv')
## Calculating Years at Company for Active Employees
active_emp_list['yearsAtCompany'] = pd.to_datetime('now') - pd.to_datetime(exited_emp_list['Most Recent Hire Date'])
active_emp_list['yearsAtCompany'] = (active_emp_list['yearsAtCompany'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)

active_emp_list['Last Promotion Date'] = np.where(active_emp_list['Last Promotion Date'].notnull(), active_emp_list['Last Promotion Date'], datetime.datetime.now())
active_emp_list['YearsSinceLastPromotion'] = pd.to_datetime(exited_emp_list['Last Promotion Date'])- pd.to_datetime(exited_emp_list['Most Recent Hire Date'])
active_emp_list['YearsSinceLastPromotion'] = (active_emp_list['YearsSinceLastPromotion'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)

## Calculating Age from Date of Birth
active_emp_list.rename(columns={'Date Of Birth':'Date of Birth'},inplace=True)
active_emp_list['Date of Birth'] = np.where(active_emp_list['Date of Birth'].notnull(), active_emp_list['Date of Birth'], datetime.datetime.now())
active_emp_list['Age'] = pd.to_datetime('now') - pd.to_datetime(active_emp_list['Date of Birth'])
active_emp_list['Age'] = (active_emp_list['Age'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)

## RenameFew Columns:
exited_emp_list.rename(columns={'Experience_In_Months_x':'Experience_In_Months','Companies_Switched_x':'Companies_Switched'\
                               ,'Previous Relevant Work Experience (Y/N)_x':'Previous Relevant Work Experience (Y/N)'},inplace=True)

#total_emp_list = pd.read_csv('Total_Employee.csv')
total_emp_list = pd.concat([exited_emp_list, active_emp_list],ignore_index = True)



# ## Calculating Age from Date of Birth
# total_emp_list['Date of Birth'] = np.where(total_emp_list['Date of Birth'].notnull(), total_emp_list['Date of Birth'], datetime.datetime.now())
# total_emp_list['Age'] = pd.to_datetime('now') - pd.to_datetime(total_emp_list['Date of Birth'])
# total_emp_list['Age'] = (total_emp_list['Age'] /np.timedelta64(1,'Y')).apply(np.ceil).astype(int)
```

```python
total_emp_list.to_csv('Total_Employee.csv',index=False)
df = total_emp_list.copy()
df = df[df['Employee Status']!='Unpaid Leave']
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='magma')
plt.title('Correlation Heatmap', fontsize=30)
plt.show()
```

```python
size = df['Service Line'].value_counts()
labels = df['Service Line'].unique()
colors = plt.cm.Wistia(np.linspace(0,1,5))

plt.figure(figsize=(10,10))
circle = plt.Circle((0,0), radius=0.7, color='white')
plt.pie(size, colors = colors, labels = labels, shadow = True, autopct = '%.2f%%')
p = plt.gcf()
p.gca().add_artist(circle)
plt.title('Employee Segmentation w.r.t Service Line', fontsize=30)
plt.legend(bbox_to_anchor=(0.5, 0., 0.9, 0.9));
```

```python
size = df['Sub Service Line'].value_counts()
labels = df['Sub Service Line'].unique()
colors = plt.cm.Set2(np.linspace(0,1,5))

plt.figure(figsize=(10,10))
circle = plt.Circle((0,0), radius=0.7, color='white')
plt.pie(size, colors = colors, labels = labels, shadow = True, autopct = '%.2f%%')
p = plt.gcf()
p.gca().add_artist(circle)
plt.title('Percentage of Employees in various Sub Service Line', fontsize=30)
plt.legend(bbox_to_anchor=(0.5, 0., 0.9, 0.9));
```

```python
size = df['Education'].dropna().value_counts()
labels = df['Education'].dropna().unique()
colors = plt.cm.spring(np.linspace(0,1,5))

plt.figure(figsize=(10,10))
circle = plt.Circle((0,0), radius=0.7, color='white')
plt.pie(size, colors = colors, labels = labels, shadow = True, autopct = '%.2f%%')
p = plt.gcf()
p.gca().add_artist(circle)
plt.title('Percentage of Education Fields', fontsize=30)
plt.legend(bbox_to_anchor=(0.5, 0., 0.9, 0.9));
```
